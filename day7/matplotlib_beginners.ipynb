## 🚢 About the Titanic Dataset

The Titanic dataset contains information about passengers on the Titanic ship.

| Column | Meaning |
|---|---|
| `Survived` | 0 = No, 1 = Yes |
| `Pclass` | Ticket class (1, 2, 3) |
| `Sex` | Male or Female |
| `Age` | Age of the passenger |
| `Fare` | Ticket price paid |

---

## 📦 Step 1 — Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## 📂 Step 2 — Load the Dataset

```python

df = pd.read_csv("Titanic-Dataset_final.csv")
df.head()
```

---
## 📊 Chart 1 — Bar Chart

**What is a Bar Chart?**  
A bar chart shows counts or values as rectangular bars.  
It is useful for comparing categories.

**Example:** How many passengers survived vs did not survive?

```python
# Count survived vs not survived
survival_counts = df['Survived'].value_counts()
print(survival_counts)
# Plot
plt.figure(figsize=(5, 4))
plt.bar(['Not Survived', 'Survived'], survival_counts)
plt.title('Survival Count')
plt.xlabel('Status')
plt.ylabel('Number of Passengers')
plt.show()
```

**What we see:**  
More passengers did NOT survive than those who did.

---
## 📊 Chart 2 — Bar Chart (Gender Count)

Let's see how many male and female passengers were on board.

```python
# Count male and female passengers
gender_counts = df['Sex'].value_counts()
print("counts ", gender_counts)
print("index ", gender_counts.index)
print("values ", gender_counts.values)
# Plot
plt.figure(figsize=(5, 4))
plt.bar(gender_counts.index, gender_counts.values)
plt.title('Gender Count')
plt.xlabel('Gender')
plt.ylabel('Number of Passengers')
plt.show()
```

---
## 📊 Chart 3 — Histogram

**What is a Histogram?**  
A histogram shows how values are spread or distributed.  
It groups values into ranges (called **bins**).

**Example:** What was the age distribution of passengers?

```python
# Plot age distribution
plt.figure(figsize=(6, 4))
plt.hist(df['Age'], bins=10)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

**What we see:**  
Most passengers were between **20 and 40 years old**.

---
## 🥧 Chart 4 — Pie Chart

**What is a Pie Chart?**  
A pie chart shows how a whole is divided into parts.  
It is useful for showing **percentages**.

**Example:** What percentage of passengers were in each class?

```python
# Count passengers in each class
class_counts = df['PassengerClass'].value_counts()

# Plot
plt.figure(figsize=(5, 5))
plt.pie(class_counts, labels=['Class 3', 'Class 1', 'Class 2'], autopct='%1.0f%%')
plt.title('Passenger Class Distribution')
plt.show()
```

**What we see:**  
More than half of the passengers were in **Class 3** (the cheapest class).

---
## 📈 Chart 5 — Line Plot

**What is a Line Plot?**  
A line plot connects data points with a line.  
It is useful for showing **trends** over time or ordered data.

**Example:** What is the average fare paid per passenger class?

```python
# Average fare per class
avg_fare = df.groupby('PassengerClass')['Fare'].mean()

# Plot
plt.figure(figsize=(5, 4))
plt.plot([1, 2, 3], avg_fare.values, marker='o')
plt.title('Average Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')
plt.show()
```

**What we see:**  
Class 1 passengers paid the **highest fare**, while Class 3 paid the lowest.

## 🏋️ Mini Exercise

**Problem:**  
You are a data analyst. Your manager wants to know:

> *"Which passenger class had the highest survival rate?"*

**Your Task:**  
1. Calculate the average survival rate for each class using `df.groupby('Pclass')['Survived'].mean()`  
2. Plot a **bar chart** showing the survival rate per class  
3. Add a title and axis labels  

Write your code in the cell below 👇

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Titanic-Dataset_final.csv')

# Convert Yes/No to 1/0
df['Survived'] = df['Survived'].map({'Yes': 1, 'No': 0})

# Group by gender and count survivors
survivors_by_gender = df.groupby('Sex')['Survived'].sum()

# Plot bar chart
plt.bar(survivors_by_gender.index, survivors_by_gender.values)

plt.title('Number of Survivors by Gender')
plt.xlabel('Gender')
plt.ylabel('Number of Survivors')

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Titanic-Dataset_final.csv')

# Plot histogram
plt.hist(df['Fare'], bins=30)

plt.title('Distribution of Ticket Fare')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Titanic-Dataset_final.csv')

# Count survivors and non-survivors
survival_counts = df['Survived'].value_counts()

# Plot pie chart
plt.pie(
    survival_counts,
    labels=['Did Not Survive', 'Survived'],
    autopct='%1.1f%%'
)

plt.title('Survival Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Titanic-Dataset_final.csv')

# Count passengers in each class
class_counts = df['PassengerClass'].value_counts()

# Plot bar chart
class_counts.plot(kind='bar')

plt.title('Number of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.show()
```

## ✅ Summary

Here is what you learned in this notebook:

| Chart | When to Use |
|---|---|
| **Bar Chart** | Compare categories (e.g. gender, class) |
| **Histogram** | See how values are spread (e.g. age, fare) |
| **Pie Chart** | Show percentages of a whole |
| **Line Plot** | Show trends across ordered data |

### Key matplotlib functions used:
- `plt.bar()` → Bar chart
- `plt.hist()` → Histogram
- `plt.pie()` → Pie chart
- `plt.plot()` → Line plot
- `plt.title()`, `plt.xlabel()`, `plt.ylabel()` → Labels
- `plt.show()` → Display the chart
