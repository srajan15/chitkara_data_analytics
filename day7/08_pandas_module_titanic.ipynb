# Module 8: GroupBy in Pandas
### 🚢 Using the Kaggle Titanic Dataset

## 1️⃣ Introduction

In data analysis, we often need to **analyze data by categories or groups**.

For example on the Titanic:

- What is the **average fare by passenger class**?
- What is the **survival rate by gender**?
- How many passengers **embarked from each port**?

To perform this type of analysis, Pandas provides a powerful function called **`groupby()`**.

The **groupby operation** follows three steps:

```
Split → Apply → Combine
```

1️⃣ **Split** the data into groups  
2️⃣ **Apply** calculations to each group  
3️⃣ **Combine** the results into a summary

## 2️⃣ Why This Concept is Important

The Titanic dataset has several categorical columns perfect for groupby analysis:

| Column     | Categories                        |
| ---------- | --------------------------------- |
| `Pclass`   | 1st class, 2nd class, 3rd class   |
| `Sex`      | male, female                      |
| `Embarked` | C (Cherbourg), Q (Queenstown), S (Southampton) |
| `Survived` | 0 (No), 1 (Yes)                   |

Questions we can answer with `groupby()`:

✔ Which **passenger class** had the highest survival rate?  
✔ What was the **average fare per class**?  
✔ How many passengers **per gender** survived?  
✔ What was the **average age by embarkation port**?

## 3️⃣ Load and Prepare the Titanic Dataset

````python
import pandas as pd

# Load the Titanic dataset
# url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv("Titanic-Dataset.csv")

# Fill missing values before groupby analysis
df["Age"]      = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

print("Shape:", df.shape)
df[["Name", "Pclass", "Sex", "Age", "Fare", "Survived", "Embarked"]].head()
````

## 4️⃣ Basic `groupby()` Operation

### Syntax
```python
df.groupby("column")["value_column"].aggregation()
```

### Titanic Example — Average Fare by Passenger Class

````python
# Average fare paid by each passenger class
df.groupby("Pclass")["Fare"].mean()
````

````python
# Average age by passenger class
df.groupby("Pclass")["Age"].mean()
````

## 5️⃣ Grouping and Calculating Total

### Titanic Example — Total Fare Collected by Passenger Class

````python
# Total fare collected from each passenger class
df.groupby("Pclass")["Fare"].sum()
````

````python
# Total number of survivors by gender
df.groupby("Sex")["Survived"].sum()
````

## 6️⃣ Counting Values in Each Group

### Titanic Example — Passenger Count per Class & per Port

````python
# Count number of passengers in each class
df.groupby("Pclass")["PassengerId"].count()
````

````python
# Count passengers who boarded from each port
df.groupby("Embarked")["PassengerId"].count()
````

````python
# Count all columns per group (full groupby count)
df.groupby("Pclass").count()
````

## 7️⃣ Multiple Aggregations with `agg()`

We can calculate **multiple statistics at once** using `.agg()`.

### Titanic Example — Fare Statistics by Passenger Class

````python
# Multiple aggregations on Fare grouped by Pclass
df.groupby("Pclass")["Fare"].agg(["mean", "max", "min", "median", "count"])
````

````python
# Multiple aggregations on Age grouped by Sex
df.groupby("Sex")["Age"].agg(["mean", "max", "min", "count"])
````

````python
# Survival rate (mean of 0/1) and total survivors grouped by passenger class
df.groupby("Pclass")["Survived"].agg(["mean", "sum"]).rename(
    columns={"mean": "SurvivalRate", "sum": "TotalSurvived"}
)
````

## 8️⃣ Grouping by Multiple Columns

We can group by **more than one column** for deeper analysis.

### Titanic Example — Survival Rate by Class AND Gender

````python
# Average survival rate grouped by Pclass AND Sex
df.groupby(["Pclass", "Sex"])["Survived"].mean()
````

````python
# Average fare grouped by Pclass AND Embarked port
df.groupby(["Pclass", "Embarked"])["Fare"].mean()
````

````python
# Count of passengers grouped by Sex AND Survived
df.groupby(["Sex", "Survived"])["PassengerId"].count()
````

## 9️⃣ Viewing the GroupBy Object & Its Groups

A `groupby()` call creates a **GroupBy object** — we can inspect it.

````python
# Create a GroupBy object
grouped = df.groupby("Pclass")

# View the group keys and indices
print(grouped.groups)
````

````python
# Get a specific group — all 1st class passengers
first_class = grouped.get_group(1)
print("1st Class passengers:", len(first_class))
first_class[["Name", "Sex", "Age", "Fare"]].head()
````

## 🔟 Iterating Through Groups

We can **loop through each group** to inspect or process them individually.

````python
# Iterate through each Passenger Class group
for pclass, group in df.groupby("Pclass"):
    print(f"\n--- Passenger Class {pclass} ---")
    print(f"  Total passengers : {len(group)}")
    print(f"  Survivors        : {group['Survived'].sum()}")
    print(f"  Survival Rate    : {group['Survived'].mean()*100:.1f}%")
    print(f"  Average Fare     : £{group['Fare'].mean():.2f}")
````

## 🌍 Real-World Example — Full Titanic GroupBy Analysis

Combining multiple groupby operations to produce a comprehensive analysis report.

````python
print("========== SURVIVAL ANALYSIS BY CLASS ==========")
class_summary = df.groupby("Pclass").agg(
    TotalPassengers = ("PassengerId", "count"),
    TotalSurvived   = ("Survived", "sum"),
    SurvivalRate    = ("Survived", "mean"),
    AvgFare         = ("Fare", "mean"),
    AvgAge          = ("Age", "mean")
).round(2)
print(class_summary)
````

````python
print("========== SURVIVAL ANALYSIS BY GENDER ==========")
gender_summary = df.groupby("Sex").agg(
    TotalPassengers = ("PassengerId", "count"),
    TotalSurvived   = ("Survived", "sum"),
    SurvivalRate    = ("Survived", "mean"),
    AvgFare         = ("Fare", "mean")
).round(2)
print(gender_summary)
````

````python
print("========== SURVIVAL ANALYSIS BY CLASS & GENDER ==========")
combined = df.groupby(["Pclass", "Sex"])["Survived"].agg(
    Count="count",
    Survived="sum",
    SurvivalRate="mean"
).round(3)
print(combined)
````

## 💡 Important Tips

✔ `groupby()` is one of the **most powerful functions in Pandas** — mastering it is essential.

✔ Always combine `groupby()` with aggregation functions:
```
sum()  mean()  count()  max()  min()  agg()
```

✔ Use **named aggregations** with `agg()` for cleaner, readable output:
```python
df.groupby("col").agg(NewName=("col", "func"))
```

✔ Use `get_group()` to extract a single group as a DataFrame for further analysis.
