# Module 7: Data Aggregation in Pandas
### 🚢 Using the Kaggle Titanic Dataset

## 1️⃣ Introduction

In data analysis, we often need to **summarize data** to understand patterns and trends.

For example:

- What is the **average age** of Titanic passengers?
- What is the **total fare** collected?
- What is the **maximum fare** paid by any passenger?

These types of calculations are called **Data Aggregation**.

Pandas provides several built-in functions for this:

- `sum()`
- `mean()`
- `min()`
- `max()`
- `count()`
- `median()`
- `std()`

These functions help us quickly understand **data distribution and patterns**.

## 2️⃣ Why This Concept is Important

Using the Titanic dataset, aggregation helps answer real business questions:

✔ What was the **average ticket fare**?  
✔ How many passengers **survived**?  
✔ What was the **oldest and youngest** passenger's age?  
✔ How **spread out** were the ticket prices?  

These insights help analysts **make data-driven decisions**.

## 3️⃣ Load and Prepare the Titanic Dataset

```python
import pandas as pd

# Load the Titanic dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

# Fill missing Age values so aggregation works cleanly
df["Age"] = df["Age"].fillna(df["Age"].mean())

print("Shape:", df.shape)
df[["Name", "Age", "Fare", "Survived", "Pclass"]].head()
```

## 4️⃣ `sum()` Function

### Purpose

Calculates the **total sum of values** in a column.

### Titanic Example — Total Fare Collected

```python
# Total fare collected from all passengers
total_fare = df["Fare"].sum()
print(f"Total Fare Collected: £{total_fare:,.2f}")
```

```python
# Total number of passengers who survived
total_survived = df["Survived"].sum()
print(f"Total Passengers Who Survived: {total_survived}")
```

## 5️⃣ `mean()` Function

### Purpose

Calculates the **average value** of a column.

Formula: `Mean = Total Values / Number of Values`

### Titanic Example — Average Age & Fare

```python
# Average age of all passengers
avg_age = df["Age"].mean()
print(f"Average Passenger Age: {avg_age:.2f} years")
```

```python
# Average fare paid
avg_fare = df["Fare"].mean()
print(f"Average Ticket Fare: £{avg_fare:.2f}")
```

```python
# Overall survival rate (mean of 0s and 1s = percentage)
survival_rate = df["Survived"].mean() * 100
print(f"Overall Survival Rate: {survival_rate:.1f}%")
```

## 6️⃣ `min()` Function

### Purpose

Finds the **minimum (smallest) value** in a column.

### Titanic Example — Youngest Passenger & Lowest Fare

```python
# Youngest passenger age
min_age = df["Age"].min()
print(f"Youngest Passenger Age: {min_age} years")
```

```python
# Lowest ticket fare paid
min_fare = df["Fare"].min()
print(f"Lowest Ticket Fare: £{min_fare:.2f}")
```

## 7️⃣ `max()` Function

### Purpose

Finds the **maximum (largest) value** in a column.

### Titanic Example — Oldest Passenger & Highest Fare

```python
# Oldest passenger age
max_age = df["Age"].max()
print(f"Oldest Passenger Age: {max_age} years")
```

```python
# Highest ticket fare paid
max_fare = df["Fare"].max()
print(f"Highest Ticket Fare: £{max_fare:.2f}")

# Find the passenger who paid the most
top_payer = df[df["Fare"] == max_fare][["Name", "Pclass", "Fare"]]
print(top_payer)
```

## 8️⃣ `count()` Function

### Purpose

Counts the **number of non-null values** in a column.

### Titanic Example — Passenger Counts

```python
# Total number of passengers with recorded fares
fare_count = df["Fare"].count()
print(f"Passengers with Fare records: {fare_count}")
```

```python
# Count passengers by gender
print(df["Sex"].value_counts())
```

```python
# Count passengers by Passenger Class
print(df["Pclass"].value_counts().sort_index())
```

## 9️⃣ `median()` Function

### Purpose

Finds the **middle value** when data is sorted.

The median is more **robust than the mean** when data contains extreme outliers.

### Titanic Example — Median Fare (better than mean due to outliers)

```python
# Median fare — less affected by very high fares paid by 1st-class passengers
median_fare = df["Fare"].median()
mean_fare   = df["Fare"].mean()

print(f"Median Fare: £{median_fare:.2f}")
print(f"Mean Fare:   £{mean_fare:.2f}")
print("\nNote: Mean is higher because a few 1st-class passengers paid very large fares.")
```

```python
# Median age of passengers
median_age = df["Age"].median()
print(f"Median Passenger Age: {median_age} years")
```

## 🔟 `std()` Function

### Purpose

Calculates the **standard deviation** — how spread out the values are.

- **Large std → values are far from the mean** (high variation)
- **Small std → values are close to the mean** (low variation)

### Titanic Example — Fare Variation

```python
# Standard deviation of Fare — measures price variation
std_fare = df["Fare"].std()
print(f"Fare Standard Deviation: £{std_fare:.2f}")
print("This large value means ticket prices varied enormously across passenger classes.")
```

```python
# Standard deviation of Age
std_age = df["Age"].std()
print(f"Age Standard Deviation: {std_age:.2f} years")
```

## 1️⃣1️⃣ Applying Aggregation on Multiple Columns

We can calculate statistics for **all numerical columns at once**.

```python
# Mean of all numerical columns
df.mean(numeric_only=True)
```

```python
# Full statistical summary — describe() uses many aggregation functions internally
df[["Age", "Fare", "SibSp", "Parch"]].describe()
```

```python
# Apply multiple aggregations to a single column using .agg()
df["Fare"].agg(["sum", "mean", "min", "max", "median", "std", "count"])
```

## 🌍 Real-World Example — Titanic Insights Dashboard

Combining all aggregation functions to produce a meaningful business summary.

```python
print("========== TITANIC PASSENGER SUMMARY ==========")
print(f"Total Passengers         : {df['PassengerId'].count()}")
print(f"Total Survived           : {df['Survived'].sum()}")
print(f"Survival Rate            : {df['Survived'].mean()*100:.1f}%")
print("")
print(f"Average Age              : {df['Age'].mean():.1f} years")
print(f"Youngest Passenger       : {df['Age'].min()} years")
print(f"Oldest Passenger         : {df['Age'].max()} years")
print("")
print(f"Total Fare Collected     : £{df['Fare'].sum():,.2f}")
print(f"Average Fare             : £{df['Fare'].mean():.2f}")
print(f"Median Fare              : £{df['Fare'].median():.2f}")
print(f"Highest Fare             : £{df['Fare'].max():.2f}")
print(f"Lowest Fare              : £{df['Fare'].min():.2f}")
print(f"Fare Std Deviation       : £{df['Fare'].std():.2f}")
print("===============================================")
```

## 💡 Important Tips

✔ Always apply aggregation functions to **numeric columns** only.

✔ Use **`median()`** instead of `mean()` when your data has extreme outliers (like Titanic fares).

✔ Use **`.agg()`** to apply multiple functions at once instead of calling them one by one.

✔ `describe()` already uses many aggregation functions internally — run it first for a quick overview.

✔ Aggregation functions become even more powerful when combined with **`groupby()`** (covered in the next module).
