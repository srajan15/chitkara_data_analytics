# Module 3: Exploring DataFrames (Basic Data Inspection)

## 1️⃣ Introduction

After loading or creating a dataset in Pandas, the **first step in data analysis** is to **explore the dataset**.

This process is called **Data Inspection** or **Data Exploration**.

It helps us quickly understand:

- What the dataset looks like
- How many rows and columns it has
- What type of data is present
- Whether there are missing values

Pandas provides several **built-in functions** that allow us to explore data easily:

- `head()`
- `tail()`
- `info()`
- `describe()`
- `shape`
- `columns`
- `index`

These functions are used **daily by data analysts**.

## 2️⃣ Why This Concept is Important

In real-world data analysis, datasets can contain **thousands or millions of rows**.

For example: Customer data, Sales data, Financial transactions, Website analytics.

Before performing analysis, we must understand:

✔ Data structure  
✔ Number of records  
✔ Data types  
✔ Missing values  

Using Pandas inspection functions helps analysts **quickly understand the dataset**.

## 3️⃣ Example Dataset

Let's create a simple DataFrame that we'll use throughout this module.

````python
import pandas as pd


df = pd.read_csv("Titanic-Dataset.csv")
df
````

## 4️⃣ `head()` Function

### Purpose

Displays the **first 5 rows** of the dataset by default.

### Explanation

- Shows the **top rows of the dataset**
- Helps us quickly see the **structure of the data**

````python
# Display the first 5 rows (default)
df.head()
````

### Custom Rows

You can also specify the number of rows to display.

````python
# Display only the first 3 rows
df.head(3)
````

## 5️⃣ `tail()` Function

### Purpose

Displays the **last 5 rows** of the dataset by default.

Useful for checking the end of the dataset.

````python
# Display the last 5 rows (default)
df.tail()
````

````python
# Display only the last 2 rows
df.tail(2)
````

## 6️⃣ `info()` Function

### Purpose

Provides **information about the dataset**.

`df.info()` shows:

- Number of rows
- Number of columns
- Column names
- Data types
- Non-null values
- Memory usage

This function is **very useful for detecting missing data**.

````python
# Get detailed information about the DataFrame
df.info()
````

## 7️⃣ `describe()` Function

### Purpose

Shows **statistical summary of numerical columns**.

`describe()` calculates:

| Statistic | Meaning            |
| --------- | ------------------ |
| count     | number of values   |
| mean      | average value      |
| std       | standard deviation |
| min       | minimum value      |
| max       | maximum value      |

This helps understand **data distribution**.

````python
# Get statistical summary of numerical columns
df.describe()
````

## 8️⃣ `shape` Attribute

### Purpose

Shows the **number of rows and columns** as a tuple `(rows, columns)`.

````python
# Get the shape of the DataFrame
print(df.shape)
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
````

## 9️⃣ `columns` Attribute

### Purpose

Displays **all column names** in the DataFrame.

````python
# Get all column names
df.columns
````

## 🔟 `index` Attribute

### Purpose

Shows the **row index values**.

Row numbers start from **0 and go up to 4** in our example.

````python
# Get the row index
df.index
````

## 💡 Important Tips

✔ **Always run these commands first** when analyzing a new dataset:

```
df.head()
df.info()
df.describe()
```

✔ They help detect **data problems early**

✔ These functions are used **in almost every data analysis project**
