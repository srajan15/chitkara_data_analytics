# Module 4: Selecting Data in Pandas (Indexing & Filtering)
### 🚢 Using the Kaggle Titanic Dataset

## 1️⃣ Introduction

In data analysis, we rarely use the **entire dataset at once**.
Most of the time, we need to **select specific rows, columns, or filtered data**.

For example, on the Titanic dataset:

- Select only the **Name column**
- Select passengers with **Fare greater than 100**
- Select the **first few rows of data**
- Select rows based on **conditions** (e.g., only female passengers)

Pandas provides several powerful methods to select data:

1️⃣ Column Selection  
2️⃣ Row Selection  
3️⃣ `loc[]` (Label-based selection)  
4️⃣ `iloc[]` (Position-based selection)  
5️⃣ Boolean Filtering  

These methods are used **daily in real-world data analysis**.

## 2️⃣ Why This Concept is Important

The Titanic dataset has **891 rows and 12 columns**.

A data analyst may need to:

✔ Find passengers who **survived**  
✔ Extract **only the Name and Fare columns**  
✔ Get **specific rows** to inspect individual records  
✔ Filter **female passengers in 1st class**  

To perform these tasks efficiently, we use **Pandas selection methods**.

## 3️⃣ Load the Titanic Dataset

We load the dataset directly from a public URL — no file download needed.

```python
import pandas as pd

# Load the Titanic dataset


df = pd.read_csv("Titanic-Dataset.csv")

print("Shape:", df.shape)
df.head()
```

```python
# Check all available columns
print(df.columns.tolist())
```

## 4️⃣ Selecting Columns

### Select One Column

Use `df["column_name"]` to select a single column.
Pandas returns a **Series** object.

```python
# Select the 'Name' column — returns a Series
df["Name"].head(10)
```

```python
# Select the 'Fare' column
df["Fare"].head(10)
```

### Select Multiple Columns

Use **double brackets** `df[["col1", "col2"]]` to select multiple columns.
This returns a **DataFrame**.

```python
# Select Name and Fare columns — returns a DataFrame
df[["Name", "Fare"]].head(10)
```

```python
# Select Name, Sex, Age, and Survived columns
df[["Name", "Sex", "Age", "Survived"]].head(10)
```

## 5️⃣ Selecting Rows

Rows can be selected using **index slicing** with the format `df[start:stop]`.

The stop index is **not included**.

```python
# Select the first 5 rows (index 0 to 4)
df[0:5]
```

```python
# Select rows 10 to 14
df[10:15]
```

## 6️⃣ Using `loc[]` — Label-Based Selection

`loc[]` selects data using **row labels and column names**.

Syntax: `df.loc[row_label, column_name]`

```python
# Select the entire row at label 0 — first passenger
df.loc[0]
```

```python
# Select a specific value: Name of the passenger at row 0
df.loc[0, "Name"]
```

```python
# Select rows 0 to 4, and only Name, Sex, Age, Survived columns
df.loc[0:4, ["Name", "Sex", "Age", "Survived"]]
```

```python
# Select rows 5 to 9, and Pclass, Fare, Embarked columns
df.loc[5:9, ["Pclass", "Fare", "Embarked"]]
```

## 7️⃣ Using `iloc[]` — Position-Based Selection

`iloc[]` selects data based on **numerical index positions** (0-based).

Syntax: `df.iloc[row_position, column_position]`

```python
# Select the entire first row by position
df.iloc[0]
```

```python
# Select a specific cell: row 0, column 3 (Name column)
# 0 → first row,  3 → 4th column (Name)
print("Column positions:")
for i, col in enumerate(df.columns):
    print(f"  {i} → {col}")
```

```python
# Get the Name of the first passenger using position
df.iloc[0, 3]
```

```python
# Select first 5 rows and first 4 columns using positions
df.iloc[0:5, 0:4]
```

```python
# Select rows 10–14 and columns 2, 3, 4 (Pclass, Name, Sex)
df.iloc[10:15, 2:5]
```

### `loc[]` vs `iloc[]` — Quick Comparison

| Feature      | `loc[]`                    | `iloc[]`                    |
| ------------ | -------------------------- | --------------------------- |
| Based on     | **Labels** (column names)  | **Positions** (numbers)     |
| Row input    | Row label / index name     | Integer position (0, 1, 2)  |
| Column input | Column name string         | Integer position (0, 1, 2)  |
| Slice end    | **Inclusive**              | **Exclusive**               |

## 8️⃣ Boolean Filtering (Conditional Selection)

We can filter rows based on **conditions**.

The condition creates a **True/False (boolean) Series**, and Pandas returns only the rows where the condition is **True**.

```python
# See what the boolean condition looks like internally
(df["Fare"] > 100).head(10)
```

```python
# Filter passengers who paid Fare > 100
df[df["Fare"] > 100][["Name", "Pclass", "Fare", "Survived"]].head(10)
```

```python
# Filter only passengers who SURVIVED (Survived == 1)
df[df["Survived"] == 1][["Name", "Sex", "Age", "Pclass"]].head(10)
```

```python
# Filter only FEMALE passengers
df[df["Sex"] == "female"][["Name", "Age", "Pclass", "Survived"]].head(10)
```

```python
# Filter passengers who embarked from port 'C' (Cherbourg)
df[df["Embarked"] == "C"][["Name", "Pclass", "Fare", "Embarked"]].head(10)
```

### Multiple Conditions

Combine conditions using `&` (AND) or `|` (OR). Always wrap each condition in **parentheses**.

```python
# Filter: Female passengers who SURVIVED
df[(df["Sex"] == "female") & (df["Survived"] == 1)][["Name", "Age", "Pclass", "Fare"]].head(10)
```

```python
# Filter: 1st class passengers who paid Fare > 100
df[(df["Pclass"] == 1) & (df["Fare"] > 100)][["Name", "Pclass", "Fare", "Survived"]].head(10)
```

```python
# Filter: Passengers who were either in 1st class OR survived
df[(df["Pclass"] == 1) | (df["Survived"] == 1)][["Name", "Pclass", "Survived"]].head(10)
```

## 🌍 Real-World Example — Titanic Analysis Workflow

Combining all selection techniques in a realistic analysis scenario.

```python
# ✔ Task 1: Extract only Name and Fare columns
df[["Name", "Fare"]].head()
```

```python
# ✔ Task 2: Find passengers who did NOT survive
did_not_survive = df[df["Survived"] == 0]
print(f"Total passengers who did not survive: {len(did_not_survive)}")
did_not_survive[["Name", "Sex", "Age", "Pclass"]].head()
```

```python
# ✔ Task 3: Select a specific passenger record using loc[]
print("Passenger at index 5:")
df.loc[5, ["Name", "Sex", "Age", "Pclass", "Survived"]]
```

```python
# ✔ Task 4: Young female survivors (Age < 18, Female, Survived)
young_female_survivors = df[
    (df["Sex"] == "female") &
    (df["Age"] < 18) &
    (df["Survived"] == 1)
]
print(f"Young female survivors: {len(young_female_survivors)}")
young_female_survivors[["Name", "Age", "Pclass", "Fare"]]
```

## 💡 Important Tips

| Syntax                              | Purpose                              |
| ----------------------------------- | ------------------------------------ |
| `df["col"]`                         | Select single column (Series)        |
| `df[["col1", "col2"]]`              | Select multiple columns (DataFrame)  |
| `df[0:5]`                           | Select rows by slice                 |
| `df.loc[row, col]`                  | Label-based row/column selection     |
| `df.iloc[row, col]`                 | Position-based selection             |
| `df[df["col"] > value]`             | Boolean / conditional filtering      |
| `df[(cond1) & (cond2)]`             | Multiple conditions — AND            |
| `df[(cond1) \| (cond2)]`            | Multiple conditions — OR             |
