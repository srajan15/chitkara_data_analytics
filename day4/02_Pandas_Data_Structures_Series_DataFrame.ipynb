# Module 2: Pandas Data Structures (Series & DataFrame)

## 1️⃣ Introduction
In Pandas, data is stored using **two main data structures**:

1️⃣ **Series**  
2️⃣ **DataFrame**

These structures allow us to store and analyze data in a **structured format**, similar to Excel tables or database tables.

Understanding these two structures is very important because almost every Pandas operation works on them.

## 2️⃣ Why This Concept is Important

In real-world data analysis:

- Data usually comes in **table format**
- Each column represents a **feature**
- Each row represents a **record**

Example dataset:

| Name | Age | City |
|------|-----|------|
| Amit | 22 | Delhi |
| Ravi | 25 | Mumbai |
| Neha | 21 | Jaipur |

To work with this data in Python, we use **Pandas DataFrame**.

If we work with only **one column**, we use **Series**.

## 3️⃣ Series

A **Series** is a **one-dimensional labeled array**.

It contains:

- Data
- Index

Example structure:

Index   Value  
0       10  
1       20  
2       30  
3       40

### Creating a Series

We can create a Series using:

- Lists
- NumPy arrays
- Dictionaries

```python
import pandas as pd

data = [10, 20, 30, 40]

s = pd.Series(data)

print(s)
```

### Output

0    10  
1    20  
2    30  
3    40  
dtype: int64

Explanation:

- 0,1,2,3 → index
- 10,20,30,40 → values
- dtype → data type

## Custom Index in Series

```python
data = [100, 200, 300]

s = pd.Series(data, index=["A", "B", "C"])

print(s["C"])
```

Output:

A    100  
B    200  
C    300

Here:
- A, B, C are **custom indexes**

## Accessing Elements in Series

```python
print(s["A"])
```

Output:

100

This retrieves the value associated with index **A**.

# 5️⃣ DataFrame

A **DataFrame** is a **two-dimensional data structure**.

It contains:

- Rows
- Columns
- Index

A DataFrame looks like a **table or spreadsheet**.

Example:

| Name | Age | City |
|------|-----|------|
| Amit | 22 | Delhi |
| Ravi | 25 | Mumbai |
| Neha | 21 | Jaipur |

## Creating a DataFrame

A DataFrame can be created using:

- Dictionary
- List
- NumPy array
- CSV or Excel file

```python
data = {
    "Name": ["Amit", "Ravi", "Neha"],
    "Age": [22, 25, 21],
    "City": ["Delhi", "Mumbai", "Jaipur"]
}

df = pd.DataFrame(data)
df["Age"]
print(df)
```

Output:

    Name   Age     City
0   Amit   22     Delhi
1   Ravi   25     Mumbai
2   Neha   21     Jaipur

Explanation:

- Columns → Name, Age, City
- Rows → individual records
- Index → 0,1,2

## Accessing Data in DataFrame

```python
print(df["Name"])
```

Output:

0    Amit  
1    Ravi  
2    Neha

This returns a **Series**.

### Access Multiple Columns

```python
print(df[["Name", "City"]])
```

Output:

    Name     City
0   Amit     Delhi
1   Ravi     Mumbai
2   Neha     Jaipur

## Real-World Example

Suppose you work in a **bank** and have customer data:

| Customer | Balance | City |
|----------|---------|------|
| Amit | 50000 | Delhi |
| Ravi | 30000 | Mumbai |
| Neha | 45000 | Jaipur |

Using Pandas you can:

- Find average balance
- Filter customers with balance > 40000
- Analyze customer distribution by city

```python
bank_data = {
    "Customer": ["Amit", "Ravi", "Neha"],
    "Balance": [50000, 30000, 45000],
    "City": ["Delhi", "Mumbai", "Jaipur"]
}

bank_df = pd.DataFrame(bank_data)

print(bank_df)
```

✔ Most real-world datasets are stored as **DataFrames**  

✔ A **column inside a DataFrame is actually a Series**
