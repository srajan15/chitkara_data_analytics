# Module 5.5: Working with Files in Pandas

## 1️⃣ Introduction

In real-world data analysis, data usually comes from **external files** such as:

- CSV files
- Excel files
- JSON files
- Databases

To analyze this data using Pandas, we first need to **load the data into a DataFrame**.

Pandas provides several functions to **read and write files** easily.

### Reading Data
- `read_csv()`
- `read_excel()`
- `read_json()`

### Writing Data
- `to_csv()`
- `to_excel()`
- `to_json()`

These functions are used **daily by data analysts to import and export datasets**.

## 2️⃣ Why This Concept is Important

In real-world projects, data is usually stored in files like:

- Sales reports (Excel)
- Website analytics (CSV)
- API data (JSON)

A data analyst loads it using Pandas:

```python
df = pd.read_csv("sales_data.csv")
```

Now the data can be **cleaned**, **analyzed**, and **visualized**.

This is why **file handling is an essential skill in data analysis**.

## 3️⃣ Setup — Create Sample Files

Since we may not have external files available, we will **create sample files first** and then practice reading them back.

This simulates a real-world workflow.

````python
import pandas as pd

# Create a sample DataFrame to use throughout this module
data = {
    "Name": ["Amit", "Ravi", "Neha", "Priya", "Karan"],
    "Age": [22, 25, 21, 24, 23],
    "City": ["Delhi", "Mumbai", "Jaipur", "Delhi", "Pune"],
    "Salary": [40000, 50000, 35000, 45000, 42000]
}

df = pd.DataFrame(data)
print(df)
````

## 4️⃣ Reading CSV Files

**CSV** stands for **Comma-Separated Values** — the most commonly used data format in data analysis.

Example CSV content:
```
Name,Age,City
Amit,22,Delhi
Ravi,25,Mumbai
Neha,21,Jaipur
```

````python
# First, save our DataFrame as a CSV file
df.to_csv("data.csv", index=False)
print("CSV file saved successfully.")
````

````python
import pandas as pd

# Read the CSV file into a DataFrame
df_csv = pd.read_csv("data.csv")

print(df_csv)
````

## 5️⃣ Reading Excel Files

Pandas can read **Excel spreadsheets** (`.xlsx` format).

> **Note:** You may need to install the `openpyxl` library:
> ```
> pip install openpyxl
> ```

````python
# First, save our DataFrame as an Excel file
df.to_excel("data.xlsx", index=False)
print("Excel file saved successfully.")
````

````python
import pandas as pd

# Read the Excel file into a DataFrame
df_excel = pd.read_excel("data.xlsx")

print(df_excel)
````

## 6️⃣ Reading JSON Files

**JSON** stands for **JavaScript Object Notation** — commonly used for API data.

Example JSON content:
```json
[
  {"Name": "Amit", "Age": 22},
  {"Name": "Ravi", "Age": 25}
]
```

````python
# First, save our DataFrame as a JSON file
df.to_json("data.json", orient="records", indent=2)
print("JSON file saved successfully.")
````

````python
import pandas as pd

# Read the JSON file into a DataFrame
df_json = pd.read_json("data.json")

print(df_json)
````

## 7️⃣ Writing Data to CSV

After analyzing data, we often want to **save the results** back to a file.

````python
# Save DataFrame to CSV — includes row index numbers
df.to_csv("output_with_index.csv")
print("Saved with index.")
````

````python
# Save DataFrame to CSV — WITHOUT row index (cleaner output)
df.to_csv("output.csv", index=False)
print("Saved without index.")
````

### `index=False` — Why Use It?

By default, Pandas writes the row numbers (0, 1, 2...) as the first column.
Using `index=False` removes those row numbers — making the CSV file **clean and professional**.

````python
# Compare: read back both files to see the difference
print("With index:")
print(pd.read_csv("output_with_index.csv").head(3))

print("\nWithout index:")
print(pd.read_csv("output.csv").head(3))
````

## 8️⃣ Writing Data to Excel

Save the DataFrame as an Excel spreadsheet.

````python
# Save DataFrame to Excel
df.to_excel("output.xlsx", index=False)
print("Excel file saved successfully.")
````

## 9️⃣ Writing Data to JSON

Export the DataFrame in JSON format — useful for sending data to APIs or web applications.

````python
# Save DataFrame to JSON
df.to_json("output.json", orient="records", indent=2)
print("JSON file saved successfully.")
````

## 🌍 Real-World Example

Suppose you are a **data analyst in a retail company**.

You receive a file `sales_data.csv`. Here is the complete workflow:

````python
import pandas as pd

# Step 1: Load the dataset
# df = pd.read_csv("sales_data.csv")   # Real usage with an actual file
df = pd.read_csv("data.csv")           # Using our sample file

# Step 2: Inspect the data
print("Shape:", df.shape)
df.head()
````

````python
# Step 3: Perform analysis — e.g. find top earners
top_earners = df[df["Salary"] > 40000].sort_values("Salary", ascending=False)
print(top_earners)
````

````python
# Step 4: Save the results to a report file
top_earners.to_excel("sales_report.xlsx", index=False)
print("Report saved as sales_report.xlsx — ready to share with management!")
````

## 💡 Important Tips

✔ **CSV is the most commonly used file format** in data analysis.

✔ Always **inspect data after loading**:
```python
df.head()
df.info()
```

✔ Use `index=False` when exporting clean reports to avoid extra index columns.

✔ Pandas supports many other formats beyond CSV, Excel, and JSON:

| Format  | Read Function       | Write Function      |
| ------- | ------------------- | ------------------- |
| CSV     | `read_csv()`        | `to_csv()`          |
| Excel   | `read_excel()`      | `to_excel()`        |
| JSON    | `read_json()`       | `to_json()`         |
| SQL     | `read_sql()`        | `to_sql()`          |
| HTML    | `read_html()`       | `to_html()`         |
| Parquet | `read_parquet()`    | `to_parquet()`      |
