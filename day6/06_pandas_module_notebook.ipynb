# Module 6: Data Manipulation in Pandas
### 🚢 Using the Kaggle Titanic Dataset

## 1️⃣ Introduction

After loading and cleaning a dataset, the next step in data analysis is **data manipulation**.

Data manipulation means **modifying the dataset** to make it suitable for analysis.

Examples of data manipulation:

- Adding new columns
- Removing unnecessary columns
- Renaming columns
- Sorting data

In this module we will learn:

- Adding columns
- Removing columns → `drop()`
- Renaming columns → `rename()`
- Sorting data → `sort_values()` and `sort_index()`

We will practise all these operations on the **real-world Kaggle Titanic dataset**.

## 2️⃣ About the Titanic Dataset

The **Titanic dataset** is one of the most famous beginner datasets in data science.

It contains information about **891 passengers** who were aboard the RMS Titanic.

Key columns:

| Column       | Description                          |
| ------------ | ------------------------------------ |
| `PassengerId`| Unique ID for each passenger         |
| `Survived`   | 0 = Did not survive, 1 = Survived    |
| `Pclass`     | Ticket class (1 = 1st, 2 = 2nd, 3 = 3rd) |
| `Name`       | Passenger name                       |
| `Sex`        | Gender                               |
| `Age`        | Age in years                         |
| `Fare`       | Ticket fare paid                     |
| `Embarked`   | Port of embarkation (C, Q, S)        |

## 3️⃣ Load the Titanic Dataset

We load the dataset directly from a public URL — no file download needed.

````python
import pandas as pd

df = pd.read_csv("Titanic-Dataset.csv")

# Quick overview
print("Shape:", df.shape)
df.head()
````

````python
# Check column names available in the dataset
print(df.columns)
````

## 4️⃣ Adding a New Column

We can easily add a new column to a DataFrame using:
```python
df["new_column_name"] = values
```

### Example: Add a `FareCategory` column

We will classify each passenger's fare as **'Low'**, **'Medium'**, or **'High'**.

````python
# Add a new 'FareCategory' column based on the Fare value
def categorize_fare(fare):
    if fare < 10:
        return "Low"
    elif fare < 50:
        return "Medium"
    else:
        return "High"

df["FareCategory"] = df["Fare"].apply(categorize_fare)

print(df[["Name", "Fare", "FareCategory"]])
````

````python
# Add a calculated column: FarePerClass = Fare divided by Pclass
df["FarePerClass"] = df["Fare"] / df["Pclass"]

print(df[["Name", "Fare", "Pclass", "FarePerClass"]].head())
````

## 5️⃣ Removing a Column

Some columns in the Titanic dataset are **not useful for analysis** (e.g., `Cabin` has too many missing values, `Ticket` is just a code).

Use `drop()` to remove them.

```
axis=1 → remove a column
axis=0 → remove a row
```

````python
# Remove the 'Cabin' column — temporary (does NOT change df)
df.drop("Cabin", axis=1).head()
````

````python
# Remove multiple unnecessary columns PERMANENTLY
df.drop(["Cabin", "Ticket"], axis=1, inplace=True)

print("Columns after dropping:", df.columns.tolist())
````

````python
# Remove a row — e.g. drop the row at index 0
df.drop(0, axis=0).head()
````

## 6️⃣ Renaming Columns

Some Titanic column names are abbreviated or unclear.
Use `rename()` to give them more descriptive names.

```python
df.rename(columns={"old_name": "new_name"})
```

````python
# Rename columns to more readable names — temporary
df.rename(columns={
    "Pclass": "PassengerClass",
    "SibSp": "SiblingsSpouses",
    "Parch": "ParentsChildren"
}).head()
````

````python
# Rename PERMANENTLY using inplace=True
df.rename(columns={
    "Pclass": "PassengerClass",
    "SibSp": "SiblingsSpouses",
    "Parch": "ParentsChildren"
}, inplace=True)

print("Updated columns:", df.columns.tolist())
````

````python
df.to_csv("output.csv", index=False)
print("Saved without index.")
````

## 7️⃣ Sorting Data — `sort_values()`

Use `sort_values()` to sort the DataFrame by the values of a column.

Default order is **ascending** (lowest to highest).

````python
# Sort passengers by Age — ascending order (youngest first)
df.sort_values("Age")[["Name", "Age", "Fare"]].head(10)
````

````python
# Sort passengers by Fare — descending order (highest payers first)
df.sort_values("Fare", ascending=False)[["Name", "PassengerClass", "Fare"]].head(10)
````

### Sorting by Multiple Columns

First sort by **PassengerClass**, then by **Fare** within each class — useful for class-wise fare analysis.

````python
# Sort by PassengerClass first, then by Fare (descending) within each class
df.sort_values(
    ["PassengerClass", "Fare"],
    ascending=[True, False]
)[["Name", "PassengerClass", "Fare"]].head(10)
````

## 8️⃣ `sort_index()` Function

Use `sort_index()` to sort the DataFrame by **row index**.

This is useful after filtering operations where the index becomes out of order.

````python
# Filter first-class passengers — index will be non-sequential
first_class = df[df["PassengerClass"] == 1]
print("First 5 index values after filter:", first_class.index[:5].tolist())
````

````python
# Sort by index — ascending (default)
first_class.sort_index().head()
````

````python
# Sort by index — descending order
first_class.sort_index(ascending=False).head()
````

## 🌍 Real-World Example — Full HR-Style Workflow on Titanic

Applying all manipulation steps together in a realistic analysis workflow.

````python
import pandas as pd

# Step 1: Reload a clean copy of the dataset
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

print("Original shape:", df.shape)
````

````python
# Step 2: Remove unnecessary columns
df.drop(["Cabin", "Ticket", "PassengerId"], axis=1, inplace=True)
print("After dropping columns:", df.columns.tolist())
````

````python
# Step 3: Rename columns for clarity
df.rename(columns={
    "Pclass": "PassengerClass",
    "SibSp": "SiblingsSpouses",
    "Parch": "ParentsChildren",
    "Embarked": "Port"
}, inplace=True)
print("Renamed columns:", df.columns.tolist())
````

````python
# Step 4: Add a new calculated column — Survival label
df["SurvivalStatus"] = df["Survived"].map({0: "Did Not Survive", 1: "Survived"})
print(df[["Name", "Survived", "SurvivalStatus"]].head())
````

````python
# Step 5: Sort by Fare — highest paying passengers first
result = df.sort_values("Fare", ascending=False)
print(result[["Name", "PassengerClass", "Fare", "SurvivalStatus"]].head(10))
````

## 💡 Important Tips

| Operation              | Syntax                                             |
| ---------------------- | -------------------------------------------------- |
| Add column             | `df["col"] = values`                               |
| Calculated column      | `df["col"] = df["other"] * value`                  |
| Remove column          | `df.drop("col", axis=1)`                           |
| Remove permanently     | `df.drop("col", axis=1, inplace=True)`             |
| Rename column          | `df.rename(columns={"old": "new"})`                |
| Sort ascending         | `df.sort_values("col")`                            |
| Sort descending        | `df.sort_values("col", ascending=False)`           |
| Sort by multiple cols  | `df.sort_values(["col1", "col2"])`                 |
| Sort by index          | `df.sort_index()`                                  |
