In NumPy, the most important object is the **NumPy Array**.

# 2️⃣ Explanation of Concepts in Simple Language

## What is ndarray?

**ndarray** means **N-Dimensional Array**.

It is a container used to store **elements of the same data type**.

**Example:**
`[10 20 30 40]`

This is a **1-dimensional array**.

NumPy arrays can have multiple dimensions:

| Dimension | Example                |
| --------- | ---------------------- |
| 1D        | [1 2 3 4]              |
| 2D        | Matrix                 |
| 3D        | Collection of matrices |



---

### Example of Dimensions

1️⃣ **1D Array**
```
[1 2 3 4]
```

2️⃣ **2D Array**
```
[[1 2 3]
 [4 5 6]]
```

3️⃣ **3D Array**
```
[
 [[1 2 3],
  [4 5 6]]
]
```

# 4️⃣ Creating NumPy Arrays

NumPy provides multiple functions to create arrays.

Important functions:
- `np.array()`
- `np.zeros()`
- `np.ones()`
- `np.empty()`
- `np.arange()`
- `np.linspace()`

---

## 1. np.array()
Used to convert a **Python list into a NumPy array**.

````python
import numpy as np

# Convert a list to an array
arr = np.array([10, 20, 30, 40])

print(arr)
````

### Output
```
[10 20 30 40]
```
### Explanation
- NumPy converts the list into an **array object**.

---

## 2. np.zeros()
Creates an array filled with **0s**.

### Syntax
`np.zeros(shape)`

````python
import numpy as np

# Create an array of 5 zeros
arr = np.zeros(5)

print(arr)
````

### Output
```
[0. 0. 0. 0. 0.]
```
### Explanation
Creates **5 elements with value 0**.

---

## 3. np.ones()
Creates an array filled with **1s**.

````python
import numpy as np

# Create an array of 4 ones
arr = np.ones(4)

print(arr)
````

### Output
```
[1. 1. 1. 1.]
```

---

## 4. np.empty()
Creates an array with **random values from memory**.

````python
import numpy as np

# Create an empty array of size 3 (values will be random garbage from memory)
arr = np.empty(3)

print(arr)
````

### Explanation
It allocates memory but **does not initialize values**. Output will look like random tiny numbers (e.g., `[6.92e-310 6.92e-310 6.92e-310]`).

---

## 5. np.arange()
Creates an array with a **range of numbers**.

### Syntax
`np.arange(start, stop, step)`

````python
import numpy as np

# Create an array from 1 to 9 (10 is excluded)
arr = np.arange(1, 10)

print(arr)
````

### Output
```
[1 2 3 4 5 6 7 8 9]
```

---

## 6. np.linspace()
Creates numbers **evenly spaced between two values**.

### Syntax
`np.linspace(start, stop, number_of_values)`

````python
import numpy as np

# Generate 5 evenly spaced numbers between 1 and 10
arr = np.linspace(1, 10, 5)

print(arr)
````

### Output
```
[ 1.    3.25  5.5   7.75 10.  ]
```
### Explanation
It generates **5 numbers evenly spaced between 1 and 10**.

# 5️⃣ Array Attributes

NumPy arrays have important attributes.

| Attribute | Meaning                     |
| --------- | --------------------------- |
| shape     | size of array               |
| size      | number of elements          |
| ndim      | number of dimensions        |
| dtype     | data type                   |
| itemsize  | memory used by each element |

````python
import numpy as np

# Create a 2D array
arr = np.array([[1, 2, 3],
                [4, 5, 6]])

print("Shape:", arr.shape)
print("Size:", arr.size)
print("Dimensions:", arr.ndim)
print("Data Type:", arr.dtype)
````

### Output
```
Shape: (2, 3)
Size: 6
Dimensions: 2
Data Type: int64
```
### Explanation
- `shape` → 2 rows, 3 columns
- `size` → total elements = 6
- `ndim` → 2 dimensional array
- `dtype` → integer type

# 6️⃣ Indexing

Indexing means **accessing elements using position numbers**.
Index starts from **0**.

## 1D Array Indexing

````python
import numpy as np

arr = np.array([10, 20, 30, 40])

# Access the 1st element (Index 0)
print("Index 0:", arr[0])

# Access the 3rd element (Index 2)
print("Index 2:", arr[2])
````

### Explanation:
```
Index:  0  1  2  3
Array: [10 20 30 40]
```

---

## 2D Array Indexing

````python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

# Access element at Row 0, Column 1
print("Row 0, Col 1:", arr[0, 1])
````

### Explanation:
```
[[1 2 3]
 [4 5 6]]
```
Row = 0, Column = 1 gives us `2`.

---

# 7️⃣ Slicing

Slicing is used to **extract multiple elements from an array**.

### Syntax
`array[start:end]`

````python
import numpy as np

arr = np.array([10, 20, 30, 40, 50])

# Extract elements from index 1 up to (but not including) index 4
print(arr[1:4])
````

### Output
```
[20 30 40]
```
Explanation: Index 1 to 3 is selected.

---

### 2D Slicing Example

````python
import numpy as np

arr = np.array([[1, 2, 3],
                [4, 5, 6]])

# Select all rows (:), but only the column at index 1
print(arr[:, 1])
````

### Output
```
[2 5]
```
Explanation:
- `:` → all rows
- `1` → second column

# 8️⃣ Real-World Example

### Example — Temperature Data
Suppose we have temperature data for 5 days.
