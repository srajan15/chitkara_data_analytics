# 1️⃣ Introduction to the Topic

In many real-world applications, we need random numbers.

Examples:
- Simulating data for experiments
- Initializing weights in machine learning models
- Generating random datasets for testing
- Creating random samples

NumPy provides a powerful module called **numpy.random**.

Important functions we will learn:
- rand()
- randn()
- randint()
- choice()
- shuffle()
- seed()

These functions are widely used in Machine Learning, Data Science, and Statistics.

---

# 2️⃣ Explanation of Concepts in Simple Language

## What are Random Numbers?

Random numbers are numbers generated **without a fixed pattern**.

Example:
```text
4, 10, 2, 7, 15
```

Each time the program runs, numbers may change.

Random numbers help simulate real-world randomness.

Examples:
- Rolling a dice
- Randomly selecting students
- Generating test data

---

# 3️⃣ Syntax

First import NumPy.

```python
import numpy as np
```

Then use random functions.

```python
np.random.function_name()
```

Examples:

```python
np.random.rand()
np.random.randn()
np.random.randint()
np.random.choice()
np.random.shuffle()
np.random.seed()
```

---

# 4️⃣ Code Examples

# Example 1 — rand()

````python
import numpy as np

# Generate 3 random numbers between 0 and 1
arr = np.random.rand(3)

print(arr)
````

### Output (example)
```text
[0.23 0.75 0.11]
```

# Example 2 — rand() with matrix

````python
import numpy as np

# Generate 2x3 matrix of random numbers
arr = np.random.rand(2,3)

print(arr)
````

# Example 3 — randn()

````python
import numpy as np

# Generate numbers from normal distribution
arr = np.random.randn(5)

print(arr)
````

Values are generated around mean = 0.

# Example 4 — randint()

````python
import numpy as np

# Generate 5 random integers between 1 and 9
arr = np.random.randint(1,10,5)

print(arr)
````

# Example 5 — choice()

````python
import numpy as np

arr = np.array([10,20,30,40])

# Select a random element
result = np.random.choice(arr)

print(result)
````

# Example 6 — choice() multiple selections

````python
import numpy as np

arr = np.array([10,20,30,40])

# Select 3 random elements
result = np.random.choice(arr,3)

print(result)
````

# Example 7 — shuffle()

````python
import numpy as np

arr = np.array([1,2,3,4,5])

# Shuffle the array
np.random.shuffle(arr)

print(arr)
````

# Example 8 — seed()

````python
import numpy as np

# Set seed for reproducibility
np.random.seed(10)

print(np.random.randint(1,10,5))
````

Using the same seed produces the same random numbers every time.

---

# 5️⃣ Explanation of Output

Example:
```python
np.random.randint(1,10,5)
```

| Parameter | Meaning |
|-----------|--------|
| 1 | starting number |
| 10 | ending number (excluded) |
| 5 | number of values |

So NumPy generates **5 random numbers between 1 and 9**.

# 6️⃣ Real-World Example

### Random Student Selection

````python
import numpy as np

students = np.array(["Amit","Rahul","Priya","Neha","Arjun"])

# Select random student
selected = np.random.choice(students)

print(selected)
````

Randomly selects one student from the list.

# 7️⃣ Important Tips

### Tip 1
Use seed() when you want repeatable results.

Example:
```python
np.random.seed(42)
```

### Tip 2

| Function | Meaning |
|---------|--------|
| rand() | random numbers between 0 and 1 |
| randn() | normal distribution numbers |
| randint() | random integers |

### Tip 3
Random functions are commonly used to simulate datasets for testing ML models.

# 8️⃣ Summary

Important NumPy random functions:

- rand() → random numbers between 0 and 1
- randn() → normal distribution numbers
- randint() → random integers
- choice() → random selection from array
- shuffle() → randomly rearrange elements
- seed() → control randomness

These functions are widely used in:
- Machine learning
- Data simulation
- Experiment reproducibility
