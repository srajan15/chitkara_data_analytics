## 🚢 About the Titanic Dataset

The Titanic dataset contains information about passengers on the Titanic ship.

| Column | Meaning |
|---|---|
| `Survived` | 0 = No, 1 = Yes |
| `Pclass` | Ticket class (1, 2, 3) |
| `Sex` | Male or Female |
| `Age` | Age of the passenger |
| `Fare` | Ticket price paid |

## 📦 Step 1 — Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## 📂 Step 2 — Load the Dataset

```python
# url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"

df = pd.read_csv("Titanic-Dataset_final.csv")
df.head()
```

---
## 💡 Why Seaborn?

With **matplotlib**, you often need many lines of code to make a chart look nice.

With **seaborn**, charts look polished automatically.

Let's set a simple theme first:

```python
# This makes all charts look clean and nice
sns.set_theme(style='whitegrid')
```

---
## 📊 Chart 1 — Count Plot

**What is a Count Plot?**  
A count plot counts how many times each category appears and shows it as bars.  
It is like a bar chart, but you don't need to count manually — seaborn does it for you!

**Example:** How many passengers survived vs did not survive?

```python
plt.figure(figsize=(5, 4))
sns.countplot(x='Survived', data=df)
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.show()
```

**What we see:**  
More passengers did **NOT** survive than those who did.

---
## 📊 Chart 2 — Count Plot with Colour (Hue)

Seaborn lets you add a **colour split** using `hue`.  
This helps you compare two groups at the same time.

**Example:** How did survival differ between males and females?

```python
plt.figure(figsize=(5, 4))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival by Gender')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Number of Passengers')
plt.show()
```

**What we see:**  
More **females survived** than males.

---
## 📊 Chart 3 — Histogram

**What is a Histogram?**  
A histogram shows how values are spread across a range.  
Seaborn's `histplot` also adds a smooth curve on top automatically.

**Example:** What was the age distribution of passengers?

```python
plt.figure(figsize=(6, 4))
sns.histplot(x='Age', data=df, bins=10)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.show()
```

**What we see:**  
Most passengers were between **20 and 40 years old**.

---
## 📦 Chart 4 — Box Plot

**What is a Box Plot?**  
A box plot shows the **spread and middle value** of a group.  
It also highlights **outliers** (unusual values).

| Part of Box | Meaning |
|---|---|
| Middle line | Median (middle value) |
| Box edges | 25% to 75% of data |
| Dots outside | Outliers |

**Example:** How does age vary across passenger classes?

```python
plt.figure(figsize=(6, 4))
sns.boxplot(x='PassengerClass', y='Age', data=df)
plt.title('Age by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')
plt.show()
```

**What we see:**  
Class 1 passengers tended to be **older** than Class 3 passengers.

---
## 🔵 Chart 5 — Scatter Plot

**What is a Scatter Plot?**  
A scatter plot shows the relationship between **two number columns**.  
Each dot is one passenger.

**Example:** Is there a relationship between Age and Fare paid?

```python
plt.figure(figsize=(6, 4))
sns.scatterplot(x='Age', y='Fare', data=df)
plt.title('Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

## 🏋️ Mini Exercise

**Problem:**  
Your manager asks:

> *"Did passenger class affect survival? Show me a chart."*

**Your Task:**  
1. Create a **count plot** of `Pclass`  
2. Add `hue='Survived'` to split the bars by survival  
3. Add a title and axis labels  

Write your code in the cell below 👇

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset (example)
df = pd.read_csv('titanic-Dataset_final.csv')

# Style
sns.set(style='whitegrid')
```

```python

plt.figure(figsize=(6,4))
sns.countplot(x='PassengerClass', hue="Survived", data=df)
plt.title('Passenger Count by Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(x='Fare', data=df, bins=30, kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title('Fare vs Survival')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Fare')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='Age', y='Fare', hue='Survived', data=df)
plt.title('Age vs Fare (Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

## ✅ Summary

Here is what you learned in this notebook:

| Chart | Seaborn Function | When to Use |
|---|---|---|
| **Count Plot** | `sns.countplot()` | Count categories |
| **Histogram** | `sns.histplot()` | See value spread |
| **Box Plot** | `sns.boxplot()` | Compare groups, spot outliers |
| **Scatter Plot** | `sns.scatterplot()` | Relationship between two numbers |
| **Heatmap** | `sns.heatmap()` | Show correlations |

### Bonus tip — useful seaborn options:
- `hue='column'` → colour by a category
- `sns.set_theme()` → make charts look clean
- `annot=True` → show numbers on heatmap
